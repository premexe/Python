```python
num = 5  # Define num as the upper limit
dictionary = {}  # Initialize as an empty dictionary

for i in range(1, num + 1):
    dictionary[i] = i * i  # Update the dictionary with squared values

print(dictionary)
```

```python
numbers=[1,2,3,4,5]

dictionary={}
for i in numbers:
   dictionary[i]=i**2
print(dictionary)
```

```python
dictionary={}
sentence='Delhi is capital of India,Mumbai is Capital of Maharashtra'
word=sentence.split()
for i in word:
  if i in dictionary:
    dictionary[i]+=1
  else:
    dictionary[i]=1
print(dictionary)
```
